# Multiple lines on one plot

```python
x=[1,2,3,4]
y1=[10,20,30,40]
y2=[5,15,25,35]
plt.plot(x,y1,label="Sales 2025")
plt.plot(x,y2,label="Sales 2026")
plt.legend()
plt.show()
```

# Bar Chart

```python
categories=['A','B','C','D']
values=[25,40,30,20]
plt.bar(categories,values,color='green')
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

# Horizontal Bar Graph

```python
plt.barh(categories,values,color='orange')
```

# Histogram

```python
import numpy as np
data=np.random.randn(100)
plt.hist(data,bins=10,color='red',edgecolor='black')
plt.title("Histogram of Random Data")
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50]
y=[5,15,25,35,45]
plt.scatter(x,y,color='darkblue')
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Pie Chart

```python
labels=['A','B','C','D']
sizes=[30,25,25,20]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)  #%print
plt.title("Pie Chart Example")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)

n, bins, patches = plt.hist(data, bins=10, edgecolor='black')

colors = ['red','blue','green','orange','purple',
          'cyan','yellow','pink','gray','brown']

for patch, color in zip(patches, colors):
    patch.set_facecolor(color)

plt.title("Histogram with Different Bar Colors")
plt.show()
```

# Subplot Multiple Charts

```python
fig, ax=plt.subplots(1,2)
ax[0].bar(['A','B','C'],[10,20,15])
ax[1].plot([1,2,3],[3,6,9])
plt.show()
```

# Figure Size

```python
plt.figure(figsize=(8,5),facecolor='yellow')
plt.plot([1,2,3],[2,4,6])
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,25,30])
plt.grid(True)
plt.show()
```

# Changing Axis Limits


```python
plt.plot([1,2,3,4],[10,20,25,30])
plt.xlim(0,5)
plt.ylim(0,35)
plt.show()
```

# Annotating Points

```python
plt.scatter([1,2,3],[5,7,9])
plt.annotate("High Point",xy=(3,9),xytext=(2,10) ,arrowprops=dict(facecolor='black'))
plt.show()
```

# Adding Legends

```python
plt.plot([1,2,3],[2,4,6],label="Data A")
plt.plot([1,2,3],[3,6,9], label="Data B")
plt.legend()
plt.show()
plt.savefig("output_chart.jpeg",dpi=300)
```
