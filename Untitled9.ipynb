```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y,color='b',linestyle=':',linewidth=3)
plt.title('x vs y')
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.show()
```

```python
y1=3*x
```

```python
y2=4*x
```

```python
y1
```

```python
y2
```

```python
plt.plot(x,y1,color='y',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.title("Double line plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':')
plt.subplot(2,1,2)
plt.plot(x,y2,color='b',linestyle='-',linewidth=3)
plt.title("Double line plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
student={'bob':87,'matt':95,'sam':27,'rajesh':56,'satyam':38}
```

```python
name=list(student.keys())
values=list(student.values())
```

```python
plt.bar(name,values,color='g')
plt.title("Bar Plot")
plt.xlabel('Boys Name')
plt.ylabel('Marks Obtained')
plt.show()
```

```python
plt.barh(name,values,color='pink')
plt.title("Bar Plot")
plt.xlabel("Marks Obtained")
plt.ylabel("Name")
plt.grid()
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,0,3,7,2,8,9]
b=[7,5,6,1,6,5,4,8,6]
```

```python
plt.scatter(x,a)
plt.plot(x,a,color='r',linestyle=':',linewidth=3)
plt.title("Scatter Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.scatter(x,a,marker='*',c='b',s=100)
plt.show()
```

```python
plt.scatter(x,a,marker='*',c='g',s=100)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.title("Scatter Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.title("Scatter Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
data=[1,3,3,3,3,9,9,5,4,4,8,8,8,6,7]
plt.hist(data)
plt.title("HistoGram Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```

```python
plt.hist(data,color='g',bins=4) #size of graph its breadth
plt.title("Histogram plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
```

```python
plt.boxplot(data) #size of graph its breadth
plt.title("Box plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```
