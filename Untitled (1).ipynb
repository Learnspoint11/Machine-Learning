````python
import pickle
import joblib
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
iris=load_iris()
X,y=iris.data,iris.target
model=DecisionTreeClassifier()
model.fit(X,y)
# wb as write Pickle Save
with open('iris_model.pkl','wb') as file:
    pickle.dump(model,file)
    # joblib Save
with open('iris_model.joblib','wb') as file:
    joblib.dump(model,file)
print('Model saved as iris_model.pkl')
print('Model saved as iris_model.joblib')
````

# Load a Model using Pickle

````python
with open('iris_model.pkl','rb') as file:
    loaded_model= pickle.load(file)
    pred=loaded_model.predict([[5.1,3.5,1.4,0.2]])
    print('prediction:',pred)
    # category 0,1,2
    
````

# Bonus :Saving model+Scalar Together

````python
from joblib import dump,load
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
scaler=StandardScaler()
model=DecisionTreeClassifier()
pipe=Pipeline([('scaler',scaler),('model',model)])
pipe.fit(X,y)
#Save file
dump(pipe,'model_pipeline.joblib')
#load and use 
loaded_pipe=load('model_pipeline.joblib')
print(loaded_pipe.predict([[5.1,3.5,1.4,0.2]]))
````

# Chapter 40: Random Forest Classifier

## Objective
Learn how to use the Random Forest algorithm for classification problems and implement it using scikit-learn.

---

## 40.1 What is Random Forest?

Random Forest is an **ensemble learning method**. It builds multiple decision trees and merges their results to get a more accurate and stable prediction.

### Key Features:
- Reduces overfitting  
- Increases accuracy  
- Works well with both numerical and categorical data  

---

## 40.2 How It Works

- It creates multiple decision trees using random samples of the dataset.  
- Each tree gives a classification.  
- The final prediction is made by **majority voting** from all trees.  

---

## 40.3 Example Use Case

Classify whether a customer will buy a new telecom plan based on:

- Age  
- Monthly Income  
- Usage Score

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier

# Load dataset
df = pd.read_csv('plan_purchase.csv')

# Preview data
print(df.head(5))

# Split features and target
X = df.drop("Purchase", axis=1)
y = df['Purchase'].map({"No": 0, "Yes": 1})  # Convert 'Yes'/'No' to 1/0

# Identify categorical and numerical columns
categorical_features = X.select_dtypes(include='object').columns
numeric_features = X.select_dtypes(exclude='object').columns

print("Categorical Features:", list(categorical_features))
print("Numerical Features:", list(numeric_features))

# ---------------- NUMERICAL PIPELINE ----------------
# Clean missing values using median
# NOTE: Scaling (StandardScaler/Normalizer) is NOT required for RandomForest
numerical_pipe = Pipeline([
    ('impute', SimpleImputer(strategy='median'))
    # ('scaler', StandardScaler()),   # Not needed for RandomForest
    # ('normalize', Normalizer())     # Not needed for RandomForest
])

# ---------------- CATEGORICAL PIPELINE ----------------
# Handle missing values + convert categories into numbers
categorical_pipe = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder(handle_unknown='ignore'))
])

# ---------------- COMBINE PIPELINES ----------------
# Apply transformations to respective column types
preprocessor = ColumnTransformer([
    ('num', numerical_pipe, numeric_features),
    ('cat', categorical_pipe, categorical_features)
])

# ---------------- FULL PIPELINE ----------------
# Preprocessing + Model together
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ))
])

# ---------------- DATA INSPECTION ----------------
print("Columns:", df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nDescribe:")
print(df.describe())
````

````python
# 40.7 Train/Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
print("Train class distribution:")
print(y_train.value_counts(normalize=True))
print("\nTest class distribution:")
print(y_test.value_counts(normalize=True))
````

````python
# 40.8 Train the random forest model

pipeline.fit(X_train, y_train)
````

````python


# 40.9 Predict and Evaluate
from sklearn.metrics import accuracy_score, classification_report
y_pred = pipeline.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Report: \n', classification_report(y_test, y_pred))
````

````python
# 40.10 Predict for a New Customer
new_customer= pd.DataFrame({ "Age": [30], "MonthlyIncome": [55000], "PlanType": ["Premium"], "UsageScore": [65]
})
#encoded=le.fit_transform(new_customer['PlanType']) print(encoded)
prediction= pipeline.predict(new_customer) 
probability= pipeline.predict_proba(new_customer)
result = "Yes" if prediction[0] == 1 else "No"
print("Purchase Prediction Result")
print("-"* 30)
print(f"Predicted Purchase : (result)")
print(f"Probability of Purchase (probability [0] [1]:.2%)")
````

````python
#40.11 Feature Importance
import matplotlib.pyplot as plt
# Extract Trained Model
rf_model=pipeline.named_steps['model']
# numeric features (unchanged)
num_features =numeric_features.tolist()
# categorical features (expanded by OneHot Encoder) pipeline steps
cat_features=pipeline.named_steps['preprocessing'] \
.named_transformers_['cat'] \
.named_steps['encode']\
.get_feature_names_out(categorical_features)
# final feature List
all_features= num_features+cat_features.tolist()
importances = rf_model.feature_importances_
feature_list= pd.DataFrame ({
"Feature": all_features,
"Importance": importances
}).sort_values (by="Importance", ascending=True)
print (feature_list["Feature"].tolist())
importances = rf_model.feature_importances_
plt.figure(figsize=(10, 6))
plt.barh(feature_list["Feature"], feature_list["Importance"])
plt.title('Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel("Feature")
plt.tight_layout()
````
