# Linear Regression

```python
import numpy as np
```

```python
x=np.array ([1,2,3,4,5])
y=np.array([12,18,25,27,35])
n=len(x)
```

```python
sum_x=np.sum(x)
sum_y=np.sum(y)
sum_x2=np.sum(x**2)
sum_xy=np.sum(x*y)
```

```python
print("Summation x =",sum_x)
print("Summation y =",sum_y)
print("Summation x2 =",sum_x2)
print("Summation xy =",sum_xy)

```

```python
m=(n*sum_xy-sum_x*sum_y)/(n*sum_x2-sum_x**2)
print("\n Slope(m) =",round(m,2))
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
c=y_mean-m*x_mean
print("Intercept (c)=",round(c,2))
print("\n Regression Equation : y= {:.2f}x+{:.2f}".format(m,c))
```

```python
y_pred=m*x+c
errors =y-y_pred
print("\n Predicted Values and Errors:")
print("x y_actual y_pred error")
for i in range(n):
    print(x[i]," ",y[i]," ",round(y_pred[i],2)," ",round(errors[i],2))
    
```

```python
SSE = np.sum(errors**2)
MSE = SSE/n
print("\n SSE =",round(SSE,2))
print(" \n MSE =",round(MSE,2))
```

```python
MAE = np.mean(np.abs(errors))
print("MAE =",round(MAE,2))
```

```python
SST=np.sum((y-y_mean)**2)
R2=1-(SSE/SST)
print("\n SSE =",round(SST,2))
print("\n R2=",round(R2,4))
```

```python
def predict(x_new):
    return m*x_new+c
    print("\n ----Prediction for New Input----")
x_input=float(input("Enter Study Hours:"))
y_output=predict(x_input)
print(f"Predicted marks = {y_output:.2f}")
```

# EXAMPLE

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
x=np.array([[1],[2],[3],[4],[5]])
y=np.array([12,18,25,27,35])
scaler=StandardScaler()
X_scaled =scaler.fit_transform(x)
model=LinearRegression()
model.fit(X_scaled,y)
y_pred=model.predict(X_scaled)
print("Model Parameters:")
print("Slope(m):",model.coef_[0])
print("Intercept :",model.intercept_)
print("\n Evaluation Metrics:")
print("MSE :",mean_squared_error(y,y_pred))
print("MAE :",mean_absolute_error(y,y_pred))
print("RMSE :",np.sqrt(mean_squared_error(y,y_pred)))
print("R2 :",r2_score(y,y_pred))
```
